```python
import pandas as pd
```

```python
data=pd.read_csv("onlinefoods.csv")
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["Gender"]=le.fit_transform(data["Gender"])
data["Marital Status"]=le.fit_transform(data["Marital Status"])
data["Occupation"]=le.fit_transform(data["Occupation"])
data["Monthly Income"]=le.fit_transform(data["Monthly Income"])
data["Family size"]=le.fit_transform(data["Family size"])
data["Output"]=le.fit_transform(data["Output"])
data["Unnamed: 12"]=le.fit_transform(data["Unnamed: 12"])
data["Educational Qualifications"]=le.fit_transform(data["Educational Qualifications"])

```

```python
data
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
x_pca=pca.fit_transform(data)
print(x_pca)
```

```python
columns=["pca1","pca2","pca3"]
dataframe=pd.DataFrame(x_pca,columns=columns)
```

```python
dataframe
```

```python
from sklearn.model_selection import train_test_split
x=dataframe.drop("pca3",axis=1)
y=dataframe["pca3"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
lt=LabelEncoder()
y_tr=lt.fit_transform(y_train)
y_te=lt.fit_transform(y_test)
print(y_train,"hhhhh ",y_tr)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
l=LinearDiscriminantAnalysis()
l.fit(x_train,y_tr)
y_pred=l.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_te)
```
